```python
from run import get_ts
import matplotlib.pyplot as plt
```

```python
time_series_data = get_ts()
```

```python
time = []
temp = []
tmp = []
for key in time_series_data.keys():
    tmp.append([int(key), time_series_data[key]])
tmp = sorted(tmp, key=lambda x: x[0])
for i in range(len(tmp)):
    time.append(tmp[i][0])
    temp.append(tmp[i][1])
print(tmp)
```

```python
plt.plot(time, temp)
plt.show()
```
