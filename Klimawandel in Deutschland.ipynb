```python
from run import get_ts #eigner Code zum Erstellen der Time-Series
import matplotlib.pyplot as plt #Fremder Code zum Darstellen der Time-Series
```

```python
time_series_data, rstations = get_ts() #Time_series erstellen
```

```python
time = [] #Zeitdaten
temp = [] #Temperaturdaten
data = [] #Metadaten
tmp = sorted([[int(key), time_series_data[key]] for key in time_series_data.keys()], key=lambda x: x[0])
for i in range(len(tmp)):
    time.append(tmp[i][0])
    temp.append(tmp[i][1][0])
    data.append(tmp[i][1][1])
```

```python
plt.plot(time, temp)
plt.xlabel("Jahr")
plt.ylabel("Temperatur")
plt.show()
```

```python
plt.plot(time, data)
plt.xlabel("Jahr")
plt.ylabel("Anzahl von bercksichtigten Daten")
plt.show()
```

```python
fig, ax1 = plt.subplots()
ax1.plot(time, temp)
ax1.set_xlabel('time (years)')
ax1.set_ylabel('Temp', color='b')

ax2 = ax1.twinx()
ax2.plot(time, data, 'r.')
ax2.set_ylabel('Amount of used datapoints', color='r')

fig.tight_layout()
plt.show()
```
